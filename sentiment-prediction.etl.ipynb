```python
import pandas as pd
```

```python
Test = pd.read_csv('drugsComTest_raw.tsv',sep='\t')
Test.head()
```

```python
Test.columns = ['Id','drugName','condition','review','rating','date','usefulCount'] #Rename Columns
Test.head()
```

```python
Train = pd.read_csv('drugsComTrain_raw.tsv',sep='\t')
Train.head()
```

```python
Train.columns = ['Id','drugName','condition','review','rating','date','usefulCount'] #Rename Columns
Train.head()
```

```python
df = pd.concat([Test,Train]) #Concatenate the dataframes to have a larger dataframe
print(Test.shape)
print(Train.shape)
print(df.shape)
```

```python
drug_df = df[['Id','review','rating']].copy() # Select columns of interest
drug_df.head()
```

```python
drug_df.to_csv('drug_df.csv') #Save dataFrame to csv
```
