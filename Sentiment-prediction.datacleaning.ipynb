```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

```python
drug_df['cleanReview'] = drug_df['review'].apply(lambda x: ' '.join([item for item in x.split() if item not in stopwords]))#remove stop words from reviews in other for the model to learn better
```

```python
drug_df.head()
```

```python

drug_df['cleanReview'] = drug_df['cleanReview'].apply(lambda x: x.lower()) # convert all reviews to lower case
drug_df.head()
```

```python
import string
drug_df['cleanReview'] = drug_df['cleanReview'].apply(lambda x: x.translate(str.maketrans('','',string.punctuation))) #remove punctuations
drug_df.head()
```

```python
drug_df['cleanReview'] = drug_df['cleanReview'].apply(lambda x: x.translate(str.maketrans('','','1234567890'))) #remove numbers
drug_df.head()
```

```python
drug_df.to_csv('drug_df_clean.csv')
```
